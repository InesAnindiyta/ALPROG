```python
# Kelas Barang
class Barang:
    def __init__(self, kode_barang, nama_barang, harga, jumlah_stok):
        self.kode_barang = kode_barang
        self.nama_barang = nama_barang
        self.harga = harga
        self.jumlah_stok = jumlah_stok

    def update_stok(self, jumlah):
        if not isinstance(jumlah, int):
            raise ValueError("Jumlah stok harus berupa angka.")
        if self.jumlah_stok + jumlah < 0:
            raise ValueError("Stok tidak boleh negatif.")
        self.jumlah_stok += jumlah

    def info_barang(self):
        return f"Kode: {self.kode_barang}, Nama: {self.nama_barang}, Harga: Rp {self.harga:,.2f}, Stok: {self.jumlah_stok}"

# Struktur Data
daftar_barang = []
inventaris = {}

# Fungsi Menampilkan Inventaris
def tampilkan_inventaris():
    if not daftar_barang:
        print("Inventaris kosong.")
        return
    print("Daftar Barang di Inventaris:")
    for barang in daftar_barang:
        print(barang.info_barang())

# Fungsi Mencari Barang
def cari_barang(kode):
    try:
        barang = inventaris[kode]
        print("Barang ditemukan:")
        print(barang.info_barang())
    except KeyError:
        print(f"Barang dengan kode '{kode}' tidak ditemukan.")

# Fungsi Menambahkan Barang
def tambah_barang():
    try:
        kode = input("Masukkan kode barang: ")
        if kode in inventaris:
            print("Barang dengan kode ini sudah ada.")
            return
        nama = input("Masukkan nama barang: ")
        harga = float(input("Masukkan harga barang: "))
        stok = int(input("Masukkan jumlah stok barang: "))

        barang = Barang(kode, nama, harga, stok)
        daftar_barang.append(barang)
        inventaris[kode] = barang
        print("Barang berhasil ditambahkan.")
    except ValueError as e:
        print(f"Kesalahan input: {e}")

# Fungsi Mengupdate Stok Barang
def update_stok_barang():
    try:
        kode = input("Masukkan kode barang: ")
        if kode not in inventaris:
            print(f"Barang dengan kode '{kode}' tidak ditemukan.")
            return
        jumlah = int(input("Masukkan jumlah stok yang akan ditambah/dikurangi: "))
        barang = inventaris[kode]
        barang.update_stok(jumlah)
        print("Stok berhasil diperbarui.")
    except ValueError as e:
        print(f"Kesalahan: {e}")

# Menu Program
def menu():
    while True:
        print("\n=== Program Manajemen Inventaris Barang ===")
        print("1. Tampilkan Inventaris")
        print("2. Tambah Barang Baru")
        print("3. Update Stok Barang")
        print("4. Cari Barang")
        print("5. Keluar")
        try:
            pilihan = int(input("Pilih opsi (1-5): "))
            if pilihan == 1:
                tampilkan_inventaris()
            elif pilihan == 2:
                tambah_barang()
            elif pilihan == 3:
                update_stok_barang()
            elif pilihan == 4:
                kode = input("Masukkan kode barang yang dicari: ")
                cari_barang(kode)
            elif pilihan == 5:
                print("Terima kasih telah menggunakan program.")
                break
            else:
                print("Opsi tidak valid. Pilih angka antara 1-5.")
        except ValueError:
            print("Kesalahan: Masukkan angka yang valid.")

# Menjalankan Program
menu()
```
